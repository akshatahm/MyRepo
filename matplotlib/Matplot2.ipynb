```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
cols = ['a','b','c','d','e']
data = [random.randint(1,20) for i in range(0,5)]

data
```

```python
# 1. bar graph
plt.bar(cols, data, color='skyblue')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart Example')

# Display the plot
plt.show()
```

```python
# 2. line graph
x=['DSA','AI','ML']
y=[40,50,25]
plt.plot(x, y, color='green')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Line Chart Example')

# Display the plot
plt.show()
```

```python
# 3. scatter graph
x=['DSA','AI','ML']
y=[11, 16, 14]
plt.scatter(x,y)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple scatter Chart Example')

# Display the plot
plt.show()
```

```python
# 4. historgram graph
y=[4, 3, 7,16, 14]
plt.hist(y)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple hist Chart Example')

# Display the plot
plt.show()
```
