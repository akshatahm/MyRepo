```python
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4]
y = [3, 2, 4, 1, 8]

ny = np.array(y)^2

plt.plot(x, y)
plt.plot(x, y, '*')
plt.plot(x)
plt.plot(ny)

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```
