```python
import matplotlib.pyplot as plt
import numpy as np
# Create the data
data = {
    "Show": ["Hometown", "Sita", "Cricket"],
    "Male": [100000, 50000, 40000],
    "Female": [50000, 200000, 40000]
}


plt.figure(figsize=(10, 6))
plt.title("Gender Distribution per Show")
plt.ylabel("Number of Viewers")
plt.xlabel("TV Show")
#plt.plot(data["Show"], np.array(data["Male"])+np.array(data["Female"]))
plt.bar(data["Show"], np.array(data["Male"])+np.array(data["Female"]), color='lightgreen')
plt.tight_layout()
plt.show()
```
